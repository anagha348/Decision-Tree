```python
import pandas as pd
dataset = pd.read_csv("/content/bank-direct-marketing-campaigns.csv")
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
upd=['job',	'marital',	'education',	'default',	'housing'	,'loan','contact',	'month',	'day_of_week']
dataset[upd]=dataset[upd].apply(le.fit_transform)
```

```python
X=dataset[['job',	'marital',	'education',	'default',	'housing'	,'loan','contact',	'month',	'day_of_week']]
y=dataset['y']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(random_state=0,max_depth=2.0).fit(X_train,y_train)
y_pred = dt.predict(X_test)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f} %".format(acc*100))
```

```python
import graphviz 
from sklearn import tree
dot_data = tree.export_graphviz(dt, out_file=None,filled=True) 
graph = graphviz.Source(dot_data) 
graph 
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(criterion='entropy',random_state=0,max_depth=3.0).fit(X_train,y_train)
y_pred = dt.predict(X_test)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f} %".format(acc*100))
```
